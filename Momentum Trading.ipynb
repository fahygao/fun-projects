```python
#calculate the 30 moiving for price
ICU['Price_Moving_Avg'] = ICU['Adj Close'].rolling(window = 30).mean()
ICU['Price_Moving_Avg50'] = ICU['Adj Close'].rolling(window = 50).mean()
ICU['Price_Moving_Avg200'] = ICU['Adj Close'].rolling(window = 200).mean()

ICU['Vol_Moving_Avg'] = ICU['Volume'].rolling(window = 30).mean()
```

```python
# columns to show if the Price Moving Average is higher than the curent price

ICU['Price_Lower_Than_MA'] = ICU['Price_Moving_Avg'].gt(ICU['Adj Close'])
ICU['Volume_Higher_Than_MA'] = ICU['Vol_Moving_Avg'].gt(ICU['Volume'])



# ICU['Price_Lower_Than_MA'] = ICU['Price_Moving_Avg'].gt(ICU['Adj Close'])
# ICU['Volume_Higher_Than_MA'] = ICU['Vol_Moving_Avg'].gt(ICU['Volume'])
```

```python
z = 1 
PL = 0.00
startP = float(ICU['Adj Close'].iloc[0])
print("Start Price:", startP)
endP = float(ICU['Adj Close'].iloc[len(ICU['Adj Close'])-1])
print("End Price",endP)
Return = (PL/startP)
Return_Per = "{:.2%}".format(Return)

#line by line in ICU to try and get buy/sell dates and PNL 

for index, row in ICU.iterrows():
    if row['Volume_Higher_Than_MA']: 
        if row['Price_Lower_Than_MA']:
            if z: 
                print(index, row['Adj Close'], '-BUY')
                closeAdj = row['Adj Close']
                PL -= closeAdj
                z = 0 
    else: 
        if not row['Price_Lower_Than_MA']:
            if not z: 
                print(index, row['Adj Close'], '-SELL')
                closeAdj = row['Adj Close']
                PL += closeAdj
                Return = (PL/startP)
                Return_Per =  "{:.2%}".format(Return)
                print("Total PNL $", round(PL, 2))
                print("Total Return %",Return_Per, "\n")
                z = 1 
Hold_Return = (endP - startP)
Hold_Return_Per =  "{:.2%}".format((endP - startP)/startP)

print("The return for holding start to end was $", Hold_Return, " Or ", Hold_Return_Per, ".")
```
